```
import numpy as np
import matplotlib.pyplot as plt
from time import time
from google.colab import files
import pickle
```

```
## Cargamos las loss
files.upload()
```

```
#########------Elegir modelo a evaluar particularmente------#########

n_epochs         =     60
learning_rate    =    0.1
momentum         =    0.5
 
```

```
#Cargar archivo a obj (cambiar obj por el nombre que queres)
Nombre_train_loss = "{}Epocas_{}LR_{}Momentum_train_losses.pickle".format(n_epochs,learning_rate,momentum)
with open(Nombre_train_loss, "rb") as f:
    train_losses = pickle.load(f)

Nombre_test_loss = "{}Epocas_{}LR_{}Momentum_test_losses.pickle".format(n_epochs,learning_rate,momentum)
with open(Nombre_test_loss, "rb") as f:
    test_losses = pickle.load(f)
#print(test_losses)
```

```
## TRAIN LOSS

nombre_imagen_loss = "{}Epocas_{}LR_{}Momentum_Loss_por_entrenamiento.png".format(n_epochs,learning_rate,momentum)
plt.xlabel("Nmeros de entrenamientos")
plt.ylabel("MSE Loss")
plt.title("MSE Loss en el conjunto de entrenamiento")
plt.plot(train_losses)
plt.savefig(nombre_imagen_loss)
#files.download(nombre_imagen_loss) 
```

```
## TEST LOSS

nombre_imagen_loss = "{}Epocas_{}LR_{}Momentum_Loss_promedio_por_poca.png".format(n_epochs,learning_rate,momentum)
plt.xlabel("Nmeros de pocas")
plt.xticks(range(0,n_epochs+1,10))
plt.ylabel("MSE Loss promediada")
plt.title("MSE Loss promediada del conjunto de testeo")
plt.plot(test_losses)
plt.savefig(nombre_imagen_loss)
#files.download(nombre_imagen_loss)
```

```
### Intentar hacer un programa que grafique 3 al mismo tiempo

## Network 1

n_epochs         =      3
learning_rate    =    0.1
momentum         =    0.5

Nombre_train_loss = "{}Epocas_{}LR_{}Momentum_train_losses.pickle".format(n_epochs,learning_rate,momentum)
with open(Nombre_train_loss, "rb") as f:
    train_losses1 = pickle.load(f)

Nombre_test_loss = "{}Epocas_{}LR_{}Momentum_test_losses.pickle".format(n_epochs,learning_rate,momentum)
with open(Nombre_test_loss, "rb") as f:
    test_losses1 = pickle.load(f)

## Network 2

n_epochs         =     12
learning_rate    =    0.1
momentum         =    0.5

Nombre_train_loss = "{}Epocas_{}LR_{}Momentum_train_losses.pickle".format(n_epochs,learning_rate,momentum)
with open(Nombre_train_loss, "rb") as f:
    train_losses2 = pickle.load(f)

Nombre_test_loss = "{}Epocas_{}LR_{}Momentum_test_losses.pickle".format(n_epochs,learning_rate,momentum)
with open(Nombre_test_loss, "rb") as f:
    test_losses2 = pickle.load(f)

## Network 3

n_epochs         =     60
learning_rate    =    0.1
momentum         =    0.5

Nombre_train_loss = "{}Epocas_{}LR_{}Momentum_train_losses.pickle".format(n_epochs,learning_rate,momentum)
with open(Nombre_train_loss, "rb") as f:
    train_losses3 = pickle.load(f)

Nombre_test_loss = "{}Epocas_{}LR_{}Momentum_test_losses.pickle".format(n_epochs,learning_rate,momentum)
with open(Nombre_test_loss, "rb") as f:
    test_losses3 = pickle.load(f)
```

```
###     Grafico Test Losses

nombre_imagen_loss = "Momentum_Loss_promedio_por_poca.png"
plt.figure(figsize=(20,6))
plt.xlabel("Nmeros de pocas")
plt.xticks(range(0,n_epochs+1,10))
plt.ylabel("MSE Loss promediada")
#plt.xscale("log")
plt.title("MSE Loss promediada del conjunto de testeo")
plt.plot(test_losses1,color="blue",linewidth=5)
plt.plot(test_losses2,color="orange",linewidth=3)
plt.plot(test_losses3,color="green",linewidth=1)
plt.savefig(nombre_imagen_loss)
```

```
#                Grafico Train loses

nombre_imagen_loss = "{}Epocas_{}LR_{}Momentum_Loss_por_entrenamiento.png".format(n_epochs,learning_rate,momentum)
plt.figure(figsize=(20,6))
plt.xlabel("Nmeros de entrenamientos")
#plt.xscale("log")
#plt.yscale("log")
plt.ylabel("MSE Loss")
plt.title("MSE Loss en el conjunto de entrenamiento")
plt.plot(train_losses1,color="blue", marker='o', markersize=5, linewidth=0)
plt.plot(train_losses2,color="orange",marker='o', markersize=3, linewidth=0)
plt.plot(train_losses3,color="green",marker='o',markersize=1, linewidth=0)
plt.savefig(nombre_imagen_loss)
#files.download(nombre_imagen_loss) 
```
